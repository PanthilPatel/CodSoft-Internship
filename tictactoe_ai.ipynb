```python
# tic tac toe with minimax algorithm
# referenced geeksforgeeks for minimax logic

import math

def print_board(board):
    print()
    print(board[0] + " | " + board[1] + " | " + board[2])
    print("--+---+--")
    print(board[3] + " | " + board[4] + " | " + board[5])
    print("--+---+--")
    print(board[6] + " | " + board[7] + " | " + board[8])
    print()

def check_winner(board, player):
    # all winning combinations
    wins = [
        [0,1,2],[3,4,5],[6,7,8],  # rows
        [0,3,6],[1,4,7],[2,5,8],  # cols
        [0,4,8],[2,4,6]           # diagonals
    ]
    for combo in wins:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] == player:
            return True
    return False

def get_empty_cells(board):
    return [i for i in range(9) if board[i] == " "]

def is_game_over(board):
    return check_winner(board, "X") or check_winner(board, "O") or len(get_empty_cells(board)) == 0

def minimax(board, is_maximizing, alpha, beta):
    if check_winner(board, "O"):
        return 10
    if check_winner(board, "X"):
        return -10
    if not get_empty_cells(board):
        return 0

    if is_maximizing:
        best = -math.inf
        for cell in get_empty_cells(board):
            board[cell] = "O"
            score = minimax(board, False, alpha, beta)
            board[cell] = " "
            best = max(best, score)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        best = math.inf
        for cell in get_empty_cells(board):
            board[cell] = "X"
            score = minimax(board, True, alpha, beta)
            board[cell] = " "
            best = min(best, score)
            beta = min(beta, best)
            if beta <= alpha:
                break
        return best

def ai_move(board):
    best_score = -math.inf
    move = None
    for cell in get_empty_cells(board):
        board[cell] = "O"
        score = minimax(board, False, -math.inf, math.inf)
        board[cell] = " "
        if score > best_score:
            best_score = score
            move = cell
    return move

def player_move(board):
    while True:
        try:
            pos = int(input("Enter position (1-9): ")) - 1
            if 0 <= pos <= 8 and board[pos] == " ":
                return pos
            else:
                print("Invalid! Try again")
        except:
            print("Enter a number please")

def main():
    print("=== Tic Tac Toe ===")
    print("You = X   AI = O")
    print("\nPositions:")
    print("1 | 2 | 3")
    print("4 | 5 | 6")
    print("7 | 8 | 9\n")

    board = [" "] * 9
    choice = input("Go first? (y/n): ").lower()
    player_turn = choice == 'y'

    while not is_game_over(board):
        print_board(board)
        if player_turn:
            move = player_move(board)
            board[move] = "X"
        else:
            print("AI thinking...")
            move = ai_move(board)
            board[move] = "O"
            print(f"AI chose position {move+1}")
        player_turn = not player_turn

    print_board(board)
    if check_winner(board, "X"):
        print("You win! Congrats")
    elif check_winner(board, "O"):
        print("AI wins!")
    else:
        print("It's a draw!")

if __name__ == "__main__":
    main()
```
